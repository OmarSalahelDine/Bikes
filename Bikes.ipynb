# Imports Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.3)
```

# Read the Dataset

```python
df = pd.read_csv('201902-fordgobike-tripdata.csv')
print(df.shape)
df.head()
```

# Data Wrangling

```python
df.info()
```

- there are data types incorrect 

```python
df['start_time'] = pd.to_datetime(df['start_time'] , format = '%Y %m %d %H:%M:%S' , errors = 'coerce')
df['end_time'] = pd.to_datetime(df['end_time'] , format = '%Y %m %d %H:%M:%S' , errors = 'coerce') 
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['member_birth_year' , 'member_gender'] , axis = 1 , inplace = True)
```

```python
df['start_station_id'].fillna(df['start_station_id'].median() , inplace =True)
df['start_station_name'].fillna(df['start_station_name'].mode()[0] , inplace =True)
df['end_station_id'].fillna(df['end_station_id'].median() , inplace =True)
df['end_station_name'].fillna(df['end_station_name'].mode()[0] , inplace =True)
```

```python
df.isnull().sum()
```

- No Missing all clean 

```python
sum(df.duplicated())
```

- No Duplicated Data 

```python
df.describe()
```

```python
from geopy.distance import great_circle
def dist_kilos(y):
    loc_1 = (y['start_station_latitude'],y['start_station_longitude']) 
    loc_2 = (y['end_station_latitude'],y['end_station_longitude'])
    return great_circle(loc_1,loc_2).kilometers

df['dist_in_kilos'] = df.apply(dist_kilos , axis = 1)
```

```python
df.head()
```

- we are doing here data extraction and engineering for lat and long columns to make new columns for distance in kilos and that will help us in analyzing

- Now, Dataset is clean and ready for EDA
